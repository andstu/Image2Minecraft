```python
import open3d as o3d
from open3d import geometry
from open3d import utility
from open3d import visualization
# from geometry import PointCloud
# import open3d.geometry.PointCloud as pc
import numpy as np
import copy
```

```python
mesh = o3d.io.read_triangle_mesh("objs/cub/mesh_0.obj")

mesh_box = o3d.geometry.TriangleMesh.create_box(width=1.0,
                                                    height=1.0,
                                                    depth=1.0)
```

```python
pc = geometry.PointCloud()
pc.points = utility.Vector3dVector(np.asarray(mesh.vertices))
```

```python
diameter = np.linalg.norm(np.asarray(pc.get_max_bound()) - np.asarray(pc.get_min_bound()))
```

```python
views = []

cur_view = [-1,-1,-2]
for i in range(27):
    cur_view[-1] += 1
    if cur_view[-1] > 1:
        cur_view[-2] += 1
        cur_view[-1] = -1
        if cur_view[-2] > 1:
            cur_view[-3] += 1
            cur_view[-2] = -1
    
    if cur_view == [0,0,0]:
        continue
    
    print(cur_view)
    views.append(cur_view.copy())
    
    
views = np.array(views) * diameter
```

```python
weights = np.zeros(len(pc.points))
for view in views:
#     print(view)
#     meme = copy.deepcopy(mesh_box).translate(view)
    camera = view
    radius = diameter * 100

#     print("Get all points that are visible from given view point")
    _, vis_pc_indexes = pc.hidden_point_removal(camera, radius)
    vis_pc = pc.select_by_index(vis_pc_indexes)
    distances = np.linalg.norm(view-np.array(vis_pc.points), axis=1)
    
    vis_pc_indexes = np.array(vis_pc_indexes)
    print(weights.shape, vis_pc_indexes.shape, distances.shape)
    weights[vis_pc_indexes] += distances
#     o3d.visualization.draw_geometries([mesh, vis_pc, meme])
```
